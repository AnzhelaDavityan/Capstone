```python
df["numTries"].value_counts().sort_index()
```

```python
plt.figure()
sns.barplot(
    x=df["numTries"].value_counts().sort_index().index,
    y=df["numTries"].value_counts().sort_index().values
)
plt.yscale("log")
plt.title("Number of Tries Distribution (Log Scale)")
plt.xlabel("Number of Tries")
plt.ylabel("Count (log)")
plt.show()
```

```python
df["dispEmo"].value_counts()
```

```python
plt.figure()
sns.countplot(data=df, x="dispEmo", order=df["dispEmo"].value_counts().index)
plt.title("Displayed Emotion Distribution")
plt.show()
```

```python
df["respEmo"].value_counts()
```

```python
plt.figure()
sns.countplot(data=df, x="respEmo", order=df["respEmo"].value_counts().index)
plt.title("Responded Emotion Distribution")
plt.show()
```

```python
conf_matrix = pd.crosstab(df["dispEmo"], df["respEmo"])
conf_matrix
```

```python
plt.figure(figsize=(12, 9))
plt.imshow(conf_matrix, cmap="Blues")

for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        plt.text(
            j, i,
            conf_matrix.iloc[i, j],
            ha="center", va="center",
            color="black", fontsize=10
        )

plt.xticks(range(len(conf_matrix.columns)), conf_matrix.columns, rotation=45, ha="right")
plt.yticks(range(len(conf_matrix.index)), conf_matrix.index)

plt.title("Displayed vs Responded Emotion (Confusion Matrix)")
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
df.groupby("dispEmo")["numTries"].mean().sort_values(ascending=False)
```

```python
plt.figure()
sns.barplot(
    x=df.groupby("dispEmo")["numTries"].mean().sort_values().index,
    y=df.groupby("dispEmo")["numTries"].mean().sort_values().values
)
plt.title("Average Number of Tries per Emotion")
plt.xticks(rotation=45)
plt.show()

```
